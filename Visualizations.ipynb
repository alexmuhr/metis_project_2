```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
bm_full_index = pd.read_csv('data/big-mac-full-index.csv')

bm_full_index.replace(to_replace = 'EUZ', value = 'EMU', inplace = True)

bm_columns = ['date', 'iso_a3', 'currency_code', 'name', 'dollar_price']
```

```python
latest_bmi = bm_full_index[bm_full_index['date'] == bm_full_index['date'][1273]]

# Drop Taiwan for now since we don't have worldbank data

latest_bmi.drop(latest_bmi[latest_bmi['name'] == 'Taiwan'].index, inplace = True)
latest_bmi = latest_bmi.reset_index().drop('index', axis = 1)
latest_bmi = latest_bmi[bm_columns]
```

```python
latest_bmi.sort_values('dollar_price', ascending = False, inplace = True)
```

```python
plt.figure(figsize = (15,5))
sns.set(style = 'white')
sns.set_context('talk')
sns.barplot(x = latest_bmi['iso_a3'], y = latest_bmi['dollar_price'], palette = 'autumn')
sns.despine()
plt.xlabel('')
plt.xticks(rotation = 75)
plt.ylabel('BIG MAC PRICE ($)')
plt.tight_layout()
plt.savefig('figures/BMI_2019');
```

```python
# Load dataframes of train and test sets from regression modelling notebook

full_train = pd.read_csv('data/full_train.csv')
full_test = pd.read_csv('data/full_test.csv')

# Set train column to indicate train vs test set
full_train['train'] = 1
full_test['train'] = 0

fitted_bmi = pd.concat([full_train, full_test])
```

```python
fitted_bmi.sort_values('dollar_price', inplace = True, ascending = False)
fitted_bmi.reset_index(inplace = True)
fitted_bmi.drop(columns = ['Unnamed: 0','index'], inplace = True)
```

```python
bar_color_list = []
#scatter_color_list = []
for switch in fitted_bmi['train']:
    if switch == 0:
        bar_color_list.append('y')
    else:
        bar_color_list.append('r')
```

```python
plt.figure(figsize = (15,5))
sns.set(style = 'white')
sns.set_context('talk')

sns.barplot(x = fitted_bmi['iso_a3'], y = fitted_bmi['dollar_price'], alpha = 0.6, palette = bar_color_list)
sns.scatterplot(x = fitted_bmi['iso_a3'], y = fitted_bmi['predictions'], s = 200)

sns.despine()
plt.xlabel('')
plt.xticks(rotation = 75)
plt.ylabel('BIG MAC PRICE ($)')
plt.tight_layout()
plt.savefig('figures/BMI_Predictions');
```
